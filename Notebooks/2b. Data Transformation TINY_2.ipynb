# 6. Replace the resulting NaNs with mean for each column.


```python
# #Replace NaNs in selected columns with the mean of values in the same column
#reimport the dataset
FEVS5yearT2 = pd.read_csv('FEVS5yeartempT2.csv', low_memory=False)
```

List of columns we want to impute.

```python
cols = ['Q103','Q104','Q106','Q107','Q113','Q118','Q119','Q126','Q133','Q136','Q137','Q138','Q139']
```

Create relationship between columns and the value you want to fill NA with.

```python
impute = {}

for col in cols:
    impute[col] = FEVS5yearT2[col].mean()
```

```python
impute
```

Apply fill NA.

```python
#Replace NaNs in chosen columns with mean of that column
FEVS5yearT2.fillna(value=impute, inplace=True)
```
